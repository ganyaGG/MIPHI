```python
#Выводим первые строки для ознакомления с устройством полученных фреймов lectures_df.
#В lectures_df находится информация о лекциях, тегах которые отвечают за предмет и типах
lectures_df.head()
```

```python
#Выводим первые строки для ознакомления с устройством полученных фреймов questions_df.
#Содержит номер правильного ответа, часть теста, к которой ответ относится, и  тэги
questions_df.head()
```

```python
#Выводим первые строки для ознакомления с устройством полученных фреймов trains_df.
trains_df.head()
```

```python
# Анализ процента правильных ответов
correct_answers_rate = trains_df['answered_correctly'].mean() * 100
print(f'Процент правильно отвеченных вопросов: {correct_answers_rate:.2f}%')
```

```python
# describe() также выдаёт эту цифру при умножение на 100
trains_df.describe()
```

```python
# Средннее время, затраченное на ответ
time_answers_rate = trains_df['prior_question_elapsed_time'].mean()
print(f'Среднее время затраченное на ответ: {time_answers_rate:.2f}')
```

```python
# Влияние объяснений на правильность ответа
explanation_correct = trains_df.groupby('prior_question_elapsed_time')['answered_correctly'].mean()
print(explanation_correct)
# В среднем продолжительность объяснений немного увеличевает процент правильных ответов
```

```python
# Объединение данных для анализа вопросов
merged_df = pd.merge(trains_df[trains_df['content_type_id'] == 0], questions_df, left_on='content_id', right_on='question_id')
merged_df.info()
```

```python
merged_df.head(10)
```

```python
# Анализ по тегам.
tag_correctness = merged_df.groupby('tags')['prior_question_elapsed_time'].mean()
print(tag_correctness)
```

```python
# Анализ по Part
part_correctness = merged_df.groupby('part')['prior_question_elapsed_time'].mean()
print(part_correctness)
```

```python
#Анализ затраченного времени на количество правильных ответов
tag_correctness = merged_df.groupby('correct_answer')['answered_correctly'].mean()
print(tag_correctness)
```

```python
tag_correctness = merged_df.groupby('correct_answer')['prior_question_had_explanation'].mean()
print(tag_correctness)
```

```python
# Посчитаем процент успешных ответов в объединенной таблице
all_question = merged_df['answered_correctly'].shape[0]
true_question  = merged_df[merged_df['answered_correctly'] == 1].shape[0]
print(f'Процент успешных ответов {(true_question *100)/all_question:.2f} %')
```

```python
# Анализ по просмотру лекций
merged_df[(merged_df['content_type_id'] == 0)].shape[0]
```

```python
# Проанализируем влияние наличие объяснения ответа на правильность. Наличие объяснения влияет на успех
explanation_correct =  merged_df.groupby('prior_question_had_explanation')['answered_correctly'].mean()
print(explanation_correct)
```

```python
# Анализ по количеству пакетов с вопросами на успеваемость
tag_correctness = merged_df.groupby('task_container_id')['answered_correctly'].mean()
print(tag_correctness)
```
